## Transform

Using <code>exchange_rate</code> and the `exchange_rates.csv` file find the exchange rate of USD to GBP. Write a transform function that

1.  Changes the `Market Cap (US$ Billion)` column from USD to GBP
2.  Rounds the Market Cap (US$ Billion)\` column to 3 decimal places
3.  Rename `Market Cap (US$ Billion)` to `Market Cap (GBP$ Billion)`


```python
exchange_rate_df
```

```python
def transform(bank_market_cap_1_df, exchange_rate):
    # Write your code here
    bank_market_cap_1_df["Market Cap (US$ Billion)"] = bank_market_cap_1_df["Market Cap (US$ Billion)"] * exchange_rate
    bank_market_cap_1_df["Market Cap (US$ Billion)"] = bank_market_cap_1_df["Market Cap (US$ Billion)"].round(3)
    bank_market_cap_1_df.rename(columns={"Market Cap (US$ Billion)": "Market Cap (GBP$ Billion)"}, inplace=True)
    return bank_market_cap_1_df
```

```python
transform(bank_market_cap_1_df, exchange_rate)
```
